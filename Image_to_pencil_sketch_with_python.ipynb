**Task 2 : Image to Pencil Sketch with Python**

Import Library

```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

Loading the image


```python
img=cv2.imread("/content/DSC_2294.jfif")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.title("Picture")
plt.imshow(img)
plt.show()
```

Convert to Gray Scale Image

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.title("Gray Picture")
plt.imshow(gray_img,cmap="gray")
plt.show()
```

Invert the Gray Scale Image

```python
inverted_img=255-gray_img
plt.imshow(inverted_img,cmap="gray")
plt.show()
```

Blur the Inverted Image

```python
blur_img = cv2.GaussianBlur(inverted_img,(21,21),0)
plt.imshow(blur_img,cmap="gray")
plt.show()
```

Pencil Sketch Image


```python
inverted_blur_img = 255-blur_img
pencil_sketch = cv2.divide(gray_img,inverted_blur_img,scale=256.0)
plt.title("Pencil sketch of Devdutt Padikkal")
plt.imshow(pencil_sketch,cmap="gray")
plt.show()
```
